```python
#1. Решить систему уравнений методом Гаусса:
#
#  x1 + x2 − x3 − 2x4 = 0,
# 2x1 + x2 − x3 + x4 = −2,
#  x1 + x2 − 3x3 + x4 = 4.
#
# Число переменных больше числа уравнений - система недоопределена, решений бесконечное число.
# Отыщем структуру этих решений:
# 2x1 + x2 − x3 +   x4 = −2, (2e)
#       x2 − x3 + -5x4 =  2,(2*1e-2e)
#           2x3 + -3x4 = -4 (1e-3e)
# тогда, если x4=c, то x3=1.5c-2, x2=6.5c, x1=-2-3c
#
```

```python
#2. Проверить на совместность и выяснить, сколько решений будет иметь система линейных уравнений:
#a) 
# 3x1 − x2 + x3 = 4,
# 2x1 − 5x2 − 3x3 = −17,
# x1 + x2 − x3 = 0;
# число неизвестных равно числу уравнений (3), определитель матрицы коэфф. не равен нулю: решение одно.
# по т.Кронеккера: rank(A)=rank(A~)=3:                                                    решение одно.
A=np.array([[3,-1,1],[2,-5,-3],[1,1,-1]])
f=np.array([4,-17,0])

print('a)')
print(np.linalg.det(A))
print(np.linalg.matrix_rank(A))
print(np.dot(np.linalg.inv(A),f))
# б) 
# 2x1 − 4x2 + 6x3 =  1, 
#  x1 − 2x2 + 3x3 = −2,
# 3x1 − 6x2 + 9x3 =  5;
# число неизвестных равно числу уравнений (3), определитель матрицы коэфф. равен нулю, 
# но b -не лежит в подпространстве а: решений нет.
# по т.Кронеккера: rank(A)=1 < rank(A~)=2:  система несовместна,  решений нет.

print('b)')
B=np.array([[2,-4,6,1],[1,-2, 3,-2],[3,-6,9,5]])
print(np.linalg.det(B[:,0:3]))
print(np.linalg.matrix_rank(B[:,0:3]))
print(np.linalg.matrix_rank(B[:,0:4]))


# в) 
#  x1 + 2x2 + 5x3 =  4,
# 3x1 +  x2 − 8x3 = −2
# число неизвестных больше числа уравнений, система недоопределена
# ранг матрицы коэфф. равен рангу расширенной - решений бесконечно

print('c)')
C=np.array([[1,2,5,4],[3,1,-8,-2]])
print(np.linalg.matrix_rank(C[:,0:4]))
print(np.linalg.matrix_rank(C))
```
